```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
finalDF = pd.merge(df,df_country,on='Country Code',how='left')
finalDF.head(2)
```

```python
cou_name = finalDF.Country.value_counts().index
```

```python
cou_value = finalDF.Country.value_counts().values
```

```python
#pie chart
# explode = (0,  0.4,  0)
plt.pie(cou_value[:3],labels =cou_name[:3],autopct='%1.1f%%')
```
